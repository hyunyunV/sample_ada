```python
from sklearn import svm
from Module.base import *
from Module.boost import *
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
from Module.Oncetrain import *
from sklearn.metrics import roc_auc_score, confusion_matrix, accuracy_score
data = pd.read_csv('Datas/credit.csv')
data = data.iloc[:1000,:]
```

# Multi Adaboost SVM
* SVM 결과값 받아서 임계점 찾는게 주요 문제 ovr 방식으로 찾기

```python
data['credit'].value_counts()
```

# no preprocessing

```python
X = data.iloc[:,:4]
y = data.iloc[:,4]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

# FIT

```python
clf = AdaBoostClassifier( learning_rate = 0.1)
clf.fit(X_train,y_train)
proba = clf.predict_proba(X_test) # --> 뭐 정확히 어떻게 계산하는건지는 모르겠음 다큐먼트 찾는 중 
```

# 비대칭에서 너무 몰림

```python
proba
```

# 결과를 위한 preprocessing

```python
y_target = label_binarize(y_train,[0,1,2])
train_proba = clf.predict_proba(X_train)
```

# 한개 확률로 판별해보기

```python
Tpoints = []
loopproba = train_proba[:,0]
for i in range(y_target.shape[1]-1):
    point = Find_Optimal_Cutoff(y_target[:,i],loopproba)
    Tpoints.append(point) # point 추가하고 여기서 첫번째에 해당하는 것들은 모두 제거해야함 
    indexs = y_target[:,i]!=1  # Threshold 찾을 때 주요값 아니면 뻄 
    y_target = y_target[indexs,:]
    loopproba = loopproba[indexs]
    
```

```python
Tpoints
```

```python
proba = proba[:,0]
pred = np.zeros(y_test.shape)-1
for i in range(len(y_test)):
    if(proba[i]>=Tpoints[0]):
        pred[i] = 0
    elif(proba[i]>=Tpoints[1]):
        pred[i] = 1
    else :
        pred[i] = 2
pred
```

```python
y_test_bina = label_binarize(y_test,[0,1,2])
proba = clf.predict_proba(X_test)
print(sum(y_test == pred)/200)
print(roc_auc_score(y_test_bina,proba))
```

# 판별선이 2개가 아니니깐 아래와 같이 구현하는 것은 불가능

```python
Tpoints = []
loopproba = proba
for i in range(y_target.shape[1]-1):
    point = Find_Optimal_Cutoff(y_target[:,i],loopproba[:,i])
    Tpoints.append(point) # point 추가하고 여기서 첫번째에 해당하는 것들은 모두 제거해야함 
    indexs = y_target[:,i] != 1  # Threshold 찾을 때 주요값 아니면 뻄 
    y_target = y_target[indexs,:] 
    loopproba = loopproba[indexs, : ]
    
```

```python
pred = np.zeros(y_test.shape)-1
for i in range(len(y_test)):
    if(proba[i,0]<=Tpoints[0]):
        pred[i] = 0
    elif(proba[i,1]<=Tpoints[1]):
        pred[i] = 1
    else :
        pred[i] = 2
pred 
```

```python
pd.Series(pred).value_counts()
```

```python
print(accuracy_score(y_test,pred))
print(roc_auc_score(y_test,proba,multi_class='ovr',average="macro"))
```
